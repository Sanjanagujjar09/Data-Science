```python
from matplotlib import pyplot as plt
city = ['Ahmedabad','Bangalore','Chennai','Delhi','Hyderabad','Kochi','Kolkata','Mumbai']
Number = []
for j in city:
    count = 0
    for i in df.index:
        if j==df['city'][i]:
            count=count+1
            i=i+1
        else:
            i=i+1
    Number.append(count)    

plt.bar(city,Number)
plt.xticks(city,rotation='vertical')
plt.xlabel("Cities")
plt.ylabel("No. of Theaters")
plt.show()
        
```

```python
plt.figure(figsize=(18,6))
city = ['Ahmedabad','Bangalore','Chennai','Delhi','Hyderabad','Kochi','Kolkata','Mumbai']
Number_of_Multiplexes = []
Number_of_Single_screen = []
for j in city:
    count1 = 0
    count2 = 0
    for i in df.index:
        if j==df['city'][i]:
            if 'Multiplex'==df['type'][i]:
                count1=count1+1
                i=i+1
            elif 'Single Screen'==df['type'][i]:
                count2=count2+1
                i=i+1
            else:
                i+i+1
    Number_of_Multiplexes.append(count1)
    Number_of_Single_screen.append(count2)
Number_of_Multiplexes    
Number_of_Single_screen
plt.subplot(1,2,1)
plt.bar(city,Number_of_Multiplexes)
plt.xticks(city,rotation='vertical')
plt.xlabel("Cities")
plt.ylabel("No_of_Single_screens")

plt.subplot(1,2,2)
plt.bar(city,Number_of_Single_screen)
plt.xticks(city,rotation='vertical')
plt.xlabel("Cities")
plt.ylabel("No_of_Multiplexes")
plt.show()
```

```python
Chain = df.groupby(['theatre_chain']).sum()
Chain = Chain.drop(Chain.index[9:10],axis=0)
Chain = Chain.drop(Chain.index[19:20],axis=0)
Chain
```

```python
%matplotlib inline
from scipy.stats.mstats import winsorize
plt.figure(figsize=(18,6))

plt.subplot(1,2,1)
No_of_Screens=Chain['no_screens']
plt.ylabel("No_of_Screens")
plt.title('No_of_Screens')
plt.boxplot(No_of_Screens)


plt.subplot(1,2,2)
winsorized_No_of_Screens = winsorize(Chain['no_screens'],(0,0.20))
plt.ylabel("No_of_Screens")
plt.title('Winsorized_No_of_Screens')
plt.boxplot(winsorized_No_of_Screens)

plt.show()
```

```python
%matplotlib inline
from scipy.stats.mstats import winsorize
plt.figure(figsize=(18,6))

plt.subplot(1,2,1)
Orignal_average_ticket_price=Chain['average_ticket_price']
plt.ylabel("average_ticket_price")
plt.title('Average_ticket_price')
plt.boxplot(Orignal_average_ticket_price)

plt.subplot(1,2,2)
winsorized_average_ticket_price = winsorize(Chain['average_ticket_price'],(0,0.17))
plt.ylabel("average_ticket_price")
plt.title('Winsorized_Average_ticket_price')
plt.boxplot(winsorized_average_ticket_price)

plt.show()
```

```python
%matplotlib inline
from scipy.stats.mstats import winsorize
plt.figure(figsize=(18,6))

plt.subplot(1,2,1)
total_seats=df['total_seats']
plt.boxplot(total_seats)
plt.title('total_seats')
plt.ylabel('total_seats')
plt.show()

plt.subplot(1,2,2)
winsorized_total_seats= winsorize(df['total_seats'],(0,0.05))
plt.boxplot(winsorized_total_seats)
plt.title('winsorized_total_seats')
plt.ylabel('total_seats')
plt.show()
```

```python
win_list = [winsorized_total_seats,winsorized_average_ticket_price,winsorized_No_of_Screens]

for variable in win_list:
    q3, q2 = np.percentile(variable, [75 ,25])
    iqr = q3 - q2

    min_val = q2 - (iqr*1.5)
    max_val = q3 + (iqr*1.5)
    
    print("Number of outliers after winsorization : {}".format(len(np.where((variable > max_val) | (variable < min_val))[0])))
```
